dataset 2

```python
skills2 = pd.read_excel("../data/skills_dataset.xlsx")
print(skills2.columns)
print(skills2.head())
```

```python
raw_entries = skills2.iloc[:, 0].dropna().astype(str).str.strip().tolist()
```

```python
print(raw_entries)
```

```python
import re

def split_paren_entries(raw_list):
    expanded = []
    for entry in raw_list:
        expanded.append(entry.strip())
        match = re.search(r"\(([^)]+)\)", entry)
        if match:
            expanded.append(match.group(1).strip())            
            expanded.append(re.sub(r"\s*\([^)]+\)", "", entry).strip())  
    return expanded
```

```python
missing_terms = ["jira", "kafka", "terraform", "sass", "rest api", "excel"]
```

```python
skills2_expanded = split_paren_entries(raw_entries)
skills2_clean = [s.lower().strip() for s in skills2_expanded if s.strip()]

print(len(skills))
```

```python

skills = list(set(skills + skills2_clean + missing_terms))

final_cleaned = []
for s in skills:
    if "," in s:
        final_cleaned.extend(part.strip() for part in s.split(","))
    else:
        final_cleaned.append(s)

skills = list(set(final_cleaned))
print(len(skills))
```

```python
GENERIC_JUNK = {"and", "or", "the", "of", "in", "for", "with", "rest", "to", "a", "is"}

skills = [s for s in skills if s not in GENERIC_JUNK]
print(len(skills))
```

```python
with open("data/skills_taxonomy.json", "w") as f:
    json.dump({"skills": skills, "categories": skill_to_category}, f)
```
