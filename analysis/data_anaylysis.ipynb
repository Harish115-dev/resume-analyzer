```python
import pandas as pd
import os
import json
```

```python
data = pd.read_excel("../data/Technology Skills.xlsx")
```

```python
data['Example'].nunique()
```

```python
len(data)
```

```python
skills = data['Example'].dropna().str.lower().str.strip().unique().tolist()
```

```python
print(skills)
```

```python
data.isnull().sum()
```

```python
skill_to_category = (
    data[['Example', 'Commodity Title']]
    .dropna()
    .drop_duplicates(subset='Example')
    .assign(Example=lambda d: d['Example'].str.lower().str.strip())
    .set_index('Example')['Commodity Title']
    .to_dict()
)
```

```python
print(os.getcwd())
```

```python
os.makedirs("data", exist_ok=True)

with open("data/skills_taxonomy.json", "w") as f:
    json.dump({"skills": skills, "categories": skill_to_category}, f)
```
